### 2. Tworzenie tensorów w PyTorch

Funkcja `torch.Tensor` przydziela pamięć dla żądanego tensora, ale ponownie wykorzystuje wartości, które wcześniej znajdowały się w tej pamięci. Aby bezpośrednio przypisać wartości do tensora podczas jego inicjalizacji, istnieje wiele alternatyw, między innymi:

* `torch.arange`: tworzy tensor zawierający kolejne wartości z danego zakresu
* `torch.Tensor (lista wejściowa)`: tworzy tensor z elementów podanej listy

### 4. Operacje na tensorach

```python
x=torch.rand(2, 2)
y=torch.rand(2, 2)
print(x); print(y)
```

```python
#dodawanie
z = x + y
print(z)

#lub
z = torch.add(x, y)
print(z)
```

#### Wywołanie `x1 + x2` tworzy nowy tensor zawierający sumę dwóch wejściowych tensorów. Możemy jednak używać także operacji **in-place**, które są wykonywane bezpośrednio na pamięci tensora. W takim przypadku zmieniamy wartości `x2` bez możliwości ponownego odwołania się do jego wartości sprzed operacji. Przykład pokazano poniżej:


```python
x1 = torch.rand(2, 3)
x2 = torch.rand(2, 3)
print("X1 (before)", x1)
print("X2 (before)", x2)

x2.add_(x1)
print("X1 (after)", x1)
print("X2 (after)", x2)
```

```python
#odejmowanie
z=x-y
print(z)

#lub
z = torch.sub(x,y)
print(z)

#lub
y.sub_(x)
print(y)
```

#### Kolejną często wykonywaną operacją jest zmiana kształtu tensora. Tensor o rozmiarze `(2, 3)` można przekształcić w dowolny inny kształt zawierający taką samą liczbę elementów (np. tensor o rozmiarze `(6)` lub `(3, 2)` itd.). W PyTorch operacja ta nazywa się `view`:

```python
x = torch.arange(6)
print(x)
```

```python
x = x.view(2, 3)
print(x)
```

```python
#odwrócenie tensora
x = x.permute(1, 0)
print(x)
```

Inne często używane operacje to **mnożenie macierzy**, które jest kluczowe w sieciach neuronowych. Bardzo często mamy wektor wejściowy, który jest przekształcany za pomocą uczonej macierzy wag. Istnieje kilka sposobów i funkcji do wykonywania mnożenia macierzy, z których niektóre wymieniamy poniżej:

* `torch.matmul`: Wykonuje iloczyn macierzy dla dwóch tensorów, przy czym zachowanie zależy od ich wymiarów. Jeśli oba wejścia są macierzami (2-wymiarowymi tensorami), wykonuje standardowy iloczyn macierzy. Dla tensorów o wyższych wymiarach funkcja wspiera **broadcasting**. Może być też zapisywane jako `a @ b`, podobnie jak w NumPy.
* `torch.mm`: Wykonuje iloczyn macierzy dla dwóch macierzy, ale **nie wspiera broadcastingu**.
* `torch.bmm`: Wykonuje iloczyn macierzy z obsługą wymiaru batch. Jeśli pierwszy tensor ma kształt `(b, n, m)`, a drugi `(b, m, p)`, wynikowy tensor będzie miał kształt `(b, n, p)` i jest obliczany jako `b` mnożeń macierzy dla każdej pary podmacierzy.
* `torch.einsum`: Wykonuje mnożenia macierzy i inne operacje (np. sumy iloczynów) używając **konwencji sumowania Einsteina**. 

Najczęściej używamy `torch.matmul` lub `torch.bmm`. Poniżej możemy spróbować mnożenia macierzy przy użyciu `torch.matmul`.

```python
x = torch.arange(6)
x = x.view(2, 3)
print('X = ', x)
```

```python
y = torch.arange(9).view(3, 3)
print('Y = ', y)
```

```python
z = torch.matmul(x, y)
print(z)
# sprawdź wynik ręcznie ;)
```

### 5. Indeksowanie

```python
#przeglądanie wybranych elementów tensorów
x = torch.rand(4,5)
print(x)

print(x[:, 0]) #tylko 1-sza kolumna
print(x[:, 3]) #tylko 2-ga kolumna
print(x[1, :]) #drugi wiersz
print(x[1, 2]) #pojedyńczy element
print(x[1, 2].item()) #j. w.
print(x[0:2 :]) #dwa pierwsze wiersze itd.
```
