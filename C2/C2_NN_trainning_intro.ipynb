```python
import torch
```

## Eksperyment 1

Mamy bardzo prosty model:

y = w * x

Czy potrafimy automatycznie poprawić wagę `w`?

```python
w = torch.tensor([1.5])
x = torch.tensor([2.0])

y = w * x
print("prediction:", y)
```

Jeśli prawidłowa odpowiedź wynosi 1.0 — czy model się myli?

Model musi wiedzieć **jak bardzo się pomylił**.

Użyjemy funkcji straty:

loss = (prediction − target)^2

```python
target = torch.tensor([1.0])
loss = (y - target)**2

print("loss:", loss)
```

Chcemy wiedzieć:

> Jak zmienić `w`, aby zmniejszyć loss?

Innymi słowy:

d(loss)/dw

```python
w = torch.tensor([1.5], requires_grad=True)

y = w * x
loss = (y - target)**2

loss.backward()

print("gradient:", w.grad)
```

```python
learning_rate = 0.1

with torch.no_grad():
    w -= learning_rate * w.grad

print("new weight:", w)
print("y_pred:", w * x)
```

```python
#backpropagation wiele razy
w = torch.tensor([1.5], requires_grad=True)

for step in range(5):

    y = w * x
    loss = (y - target)**2

    loss.backward()

    with torch.no_grad():
        w -= 0.1 * w.grad

    w.grad.zero_()

    print(step, "loss:", loss.item(), "w:", w.item())
```
