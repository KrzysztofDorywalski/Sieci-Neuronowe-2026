Pierwszą rzeczą, którą musimy zrobić, jest określenie, które tensory wymagają obliczania gradientów. Domyślnie, gdy tworzymy tensor, nie wymaga on obliczania gradientów.

```python
x = torch.ones((3,))
print(x)
print(x.requires_grad)
```

Możemy to zmienić dla już istniejącego tensora, używając funkcji requires_grad_() (podkreślenie oznacza, że jest to operacja wykonywana w miejscu, czyli modyfikująca obiekt bez tworzenia kopii). Alternatywnie, podczas tworzenia tensora można przekazać argument requires_grad=True do większości wcześniej używanych funkcji inicjalizujących.

```python
x.requires_grad_(True)
print(x.requires_grad)
```

```python
#inicjalizacja tensora + obliczenie gradientu
a=torch.tensor([1.0],requires_grad=True)
x=torch.tensor([-2.],requires_grad=True)
```

```python
#przykład
Q=a**2+x**3
print(Q)
```

```python
#liczymy gradient 
Q.backward()
```

```python
print("Dq/Da=",a.grad)
print("Dq/Dx=",x.grad)
```

### This will fail

```python
a=torch.tensor([1.0,2.0],requires_grad=True)
x=torch.tensor([-2.,-3.0],requires_grad=True)
Q=a**2+x**3
print(Q)
Q.backward()
print("Dq/Da=",a.grad)
print("Dq/Dx=",x.grad)
```

## This will work

```python
a=torch.tensor([1.0,2.0],requires_grad=True)
x=torch.tensor([-2.,-3.0],requires_grad=True)
Q=a**2+x**3
print(Q)
Q.backward(gradient=torch.tensor([1.0,1.0]))
print("Dq/Da=",a.grad)
print("Dq/Dx=",x.grad)
```
