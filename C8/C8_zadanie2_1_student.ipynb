**Twoje zadanie:**
1. Zmodyfikuj pierwszą warstwę splotową (`conv1`), aby przyjmowała kolorowe obrazy.
2. Przelicz ponownie, jaki będzie rozmiar tensora po przejściu przez dwie warstwy splotowe i pooling. Pamiętaj, że obraz wejściowy jest teraz większy (32x32 zamiast 28x28)!
3. Zaktualizuj liczbę wejść (`in_features`) w pierwszej warstwie liniowej (`fc1`).

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class LeNet_CIFAR(nn.Module):
    def __init__(self):
        super(LeNet_CIFAR, self).__init__()
        # TODO 1: Zmień liczbę kanałów wejściowych na odpowiednią dla obrazów RGB
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5) 
        
        self.conv2 = nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5)
        
        # TODO 2: Przelicz wymiary i zmień wartość '16 * 4 * 4' na prawidłową.
        # Wskazówka: wejście to 32x32. Po conv1 (kernel 5) -> 28x28. 
        # Po max_pool2d -> ?
        # Po conv2 (kernel 5) -> ?
        # Po drugim max_pool2d -> ?
        self.fc1 = nn.Linear(in_features=16 * 4 * 4, out_features=120)
        
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = F.max_pool2d(F.relu(self.conv1(x)), 2)
        x = F.max_pool2d(F.relu(self.conv2(x)), 2)
        
        # TODO 3: Zaktualizuj wymiar do spłaszczenia (musi być taki sam jak w fc1)
        x = x.view(-1, 16 * 4 * 4) 
        
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
# Testujemy model na losowym tensorze udającym obrazek CIFAR-10 (3 kanały, 32x32)
cifar_image = torch.randn(1, 3, 32, 32)
model_cifar = LeNet_CIFAR()

print("Próba przepuszczenia obrazu CIFAR-10 przez sieć...")
output = model_cifar(cifar_image)
print("Udało się! Kształt wyjścia:", output.shape)
```
